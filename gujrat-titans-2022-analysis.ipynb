# <p style="background-color:#1B2133;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Win/Loss At Different Venue</p>

```python
results = matches.loc[(matches['Team1']==Team_name) | (matches['Team2']==Team_name), :].copy()
results['Vs'] = results['Team1'] + results['Team2']
results['Vs'] = results['Vs'].str.replace(Team_name, '')
```

```python
results.drop(['Team1','Team2'], axis=1, inplace=True)
```

```python
results['WinningTeam'] = results['WinningTeam'].apply(lambda x: 'Win' if x==Team_name else 'Loss')
results.set_index('ID', inplace=True)
```

```python
venue_wise = results.groupby(['Venue','WinningTeam']).count()['City']
venue_wise = venue_wise.unstack()
venue_wise.sort_values(by=['Loss','Win'],ascending=[False,True],inplace=True)
venue_wise.style.background_gradient(cmap='gray_r')
```

```python
plt.figure(figsize=(35,6))
venue_wise.plot(kind='barh' , stacked=True, color=[Color1,Color2])
plt.title("Results at different venues")

plt.xlabel('Wins and losses', fontsize=20)
plt.ylabel('Venues', fontsize=20)

plt.savefig('Venue-wise results', bbox_inches='tight')
plt.show()
```

```python
pd.options.plotting.backend = "plotly"

# using Plotly Express via the Pandas backend
fig1 = venue_wise.plot.bar()
fig1.show()
```

# <p style="background-color:#1B2133;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Winning Result Win/Loss Based on Toss</p>

```python
results['Toss'] = results['TossWinner'].apply(lambda x: 'Won the toss' if x==Team_name else 'Lost the toss')
results.rename(columns={'WinningTeam':'Result'}, inplace=True)
results.drop('SuperOver', axis=1, inplace=True)
```

```python
results['First Innings'] = np.where(results['TossWinner']==Team_name,results['TossDecision'],'0'+results['TossDecision'])
```

```python
results['First Innings'] = results['First Innings'].str.replace('0field','bat')
results['First Innings'] = results['First Innings'].str.replace('0bat','field')
```

```python
results.drop('TossWinner', axis=1, inplace=True)
results.drop('TossDecision', axis=1, inplace=True)
```

```python
toss_stats = results[['Result','Toss','Season']].groupby(['Toss','Result']).count()
toss_stats.style.background_gradient(cmap='gray_r')
```

# <p style="background-color:#1B2133;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Comparing PowerPlay Score to the rest of the Team</p>

```python
over_stats = df[['overs','total_run']].groupby('overs').mean()*6
```

```python
over_stats.rename(columns={'total_run':'league_average'})
over_stats['stats'] = df.loc[df['BattingTeam']==Team_name,['overs','total_run']].groupby('overs').mean()*6
over_stats.index=np.arange(1,21)

plt.figure(figsize=(10,6))
plt.title('Batting Perfomance as comaparing to others')

plt.plot(over_stats['total_run'].index, over_stats['total_run'].values,label='Others Team Average',color='green', linestyle=':')
plt.plot(over_stats['stats'], label=Team_name,color=Color2)

plt.axvline(x=6, color=Color1, alpha=0.7, linewidth=1, label='End of Powerplay')


plt.xticks(np.arange(0,21,5))
plt.ylabel('Runs')
plt.xlabel('Overs')
plt.legend()

plt.savefig('Batting Perfomance as comaparing to others', bbox_inches='tight')
plt.show()
```

# <p style="background-color:#DBBE6E;font-family:newtimeroman;color:#FFF9ED;text-align:center;border-radius:10px 10px;"><b>Observation : </b>In the powerplay , in comparison to the Other team average GT's performance was low as you can see in the graph. But after keeping up with the other team average in the death overs, they make up for the performance in powerplay,in the death overs where they have clearly score more than others at the end of 20th over
   
 

# <p style="background-color:#1B2133;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">GT Bowling Comparing to Other Teams</p>

# <p style="background-color:#1B2133;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">GT's Most Consistent Battsman</p>


```python
top_runs = df[['batter','batsman_run']].loc[df['BattingTeam']==Team_name].groupby('batter').sum().sort_values(
    by='batsman_run',ascending=False)[0:5]
top_run_getters = top_runs.index
top_runs.style.background_gradient(cmap='gray_r')
```

```python
top5 = df[['ID','batter','batsman_run']].loc[(df['BattingTeam']==Team_name) 
                                             & (df['batter'].isin(top_run_getters))
                                            ].groupby(['ID','batter']).sum().sort_index().unstack()
top5 = top5.cumsum()
top5.style.background_gradient(cmap='gray_r')
```

```python
opponents_dates = results[['ID','Vs','Date']].copy().sort_values(by='ID')
opponents_dates['Vs'].replace({'Punjab Kings':'PK','Kolkata Knight Riders':'KKR','Rajasthan Royals':'RR',
                              'Mumbai Indians':'MI','Chennai Super Kings':'CSK','Delhi Capitals':'DC',
                              'Lucknow Super Giants':'LSG','Sunrisers Hyderabad':'SRH','Royal Challengers Bangalore':'RCB'
                              ,'Gujarat Titans':'GT'},
                              inplace=True)
opponents_dates['Index'] = '( ' + opponents_dates['Date'] + ' )  ' + opponents_dates['Vs'] 
opponents_dates.head().style.background_gradient(cmap='gray_r')
```

```python
top5.index = opponents_dates['Index'].astype('str')
```

```python
top5 = top5.fillna(method='ffill').fillna(0).applymap(lambda x: int(x))
top5.style.background_gradient(cmap='gray_r')
```

```python
plt.figure(figsize=(10,6))
plt.plot(top5[('batsman_run','DA Miller')], label='Miller', color=Color1, linewidth=2)
plt.plot(top5[('batsman_run','HH Pandya')], label='Pandya', color=Color2, linewidth=2)
plt.plot(top5[('batsman_run','R Tewatia')], label='Tewatia', color='#242F9B', linewidth=2, linestyle=':')
plt.plot(top5[('batsman_run','Shubman Gill')], label='Gill', color='#242F9B', linewidth=2, linestyle='--')
plt.plot(top5[('batsman_run','WP Saha')], label='Saha', color='#FBCB0A', linewidth=2)

plt.title("How GTs top 5 runs Scorrer season played")
plt.xticks(rotation=90)
plt.xlabel('Versus')
plt.ylabel('Total Runs')
plt.legend()
plt.grid()


plt.savefig('How GTs top 5 runs Scorrer season played', bbox_inches='tight')
plt.show()
```

# <p style="background-color:#DBBE6E;font-family:newtimeroman;color:#FFF9ED;text-align:center;border-radius:10px 10px;"><b>Observation : </b>So satisfying to see! Miller,Pandaya,Gill have been incredibly consistent with their batting . Saha not played first 5 games but still score well.

# <p style="background-color:#1B2133;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">GT's Bowling Performance</p>


```python
bowling_df = df.loc[df['ID'].isin(match_ids) ]
bowling = bowling_df.loc[(bowling_df['BattingTeam']!=Team_name) & (bowling_df['isWicketDelivery']==1) & (bowling_df['kind']!='run out')]
top_bowlers = bowling['bowler'].value_counts()[0:5].index
top_bowlers
```

```python
bowling = bowling.loc[bowling['bowler'].isin(top_bowlers)]
bowling = bowling[['ID','bowler','isWicketDelivery']].groupby(['ID','bowler']).sum().unstack()
bowling.style.background_gradient(cmap='gray_r')
```

```python
bowling = bowling.cumsum().fillna(method='ffill').fillna(0)
```

```python
bowling.index = top5.index
```

```python
bowling.style.background_gradient(cmap='gray_r')
```

```python
plt.figure(figsize=(10,6))
# plt.style.use('seaborn')
plt.plot(bowling['isWicketDelivery','HH Pandya'],color='#06283D', label='Pandya',linewidth=2)
plt.plot(bowling['isWicketDelivery','LH Ferguson'],color='#FFA500', label='Ferguson',linewidth=2)
plt.plot(bowling['isWicketDelivery','Mohammed Shami'],color='#1363DF', label='Shami',linewidth=2)
plt.plot(bowling['isWicketDelivery','Rashid Khan'],color='#1363DF',linewidth=2, label='Rashid', linestyle='--')
plt.plot(bowling['isWicketDelivery','Yash Dayal'],color='#FFA500',linestyle='--', label='Yash',linewidth=2)
plt.xlabel('versus')
plt.ylabel('Total wickets')
plt.xticks(rotation=90)
plt.title("How GT's top 5 wicket takers Played in IPL 2022")
plt.legend()
plt.grid(True)
plt.savefig('How GT top 5 wicket takers Played in IPL 2022',bbox_inches='tight')
plt.show()
```
